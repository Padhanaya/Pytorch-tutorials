#### when we drive gradient of paramters of deep neural network with respect of loss, we use a lot of differentiation which is humanly imposible. In this condition we use autograde to find gradients of the parameters w.r.t loss.

```python
import torch
x = torch.tensor(3.0, requires_grad = True)
```

```python
y = x**2
```

```python
x
```

```python
# yahan torch ko pata h ki y x ka function hai, isliye jab hum y.backward() karenge to wo x ke respect me differentiate 
# karega aur x.grad me uska result store karega.
y
```

```python
y.backward()
```

```python
x.grad
```

yahan input tensor is leaf i.e., x and output tensor is root i.e., z
gradient direction batata h slope ki sath hi wo direction bhi batata h jisse ham ye decide karte h parameter ko badhaye ya
ghataye jisse ki wo minimun loss attain kar sake

```python
# ab ham parameters ko loss ke sath 
x = torch.tensor(8.9)  # this is tensor of gpa 
y = torch.tensor(0)  # this is tensor of pass or fail
```

```python
# parameters
w = torch.tensor(0.5, requires_grad = True)
b = torch.tensor(0.0, requires_grad = True)
```

```python
# feedforward
z = w*x + b
z
```

```python
# applying activation function
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
# loss function
loss = F.binary_cross_entropy(y_pred, y.float())
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
b.grad
```

```python
# creating a full neuron
x = torch.tensor([1, 2, 3, 2])
y = torch.tensor([0, 0, 1, 0], dtype = torch.float32)
```

```python
w = torch.tensor(0.6, requires_grad = True)
b = torch.tensor(0.0, requires_grad = True)
```

```python
z = w*x + b
z
```

```python
loss = F.binary_cross_entropy_with_logits(z, y)
loss
```

```python
loss.backward()
```

```python
# It show the direction and magnitude of change required in w to reduce the loss
w.grad
```

```python
b.grad
```

```python
# lekin agar ham above codes ko dobara chalaye to w.grad aur b.grad me pehle wale gradients add ho jayenge, isliye hamko 
# unko zero karna padega.
w.grad.zero_()
b.grad.zero_()
```

```python
# backward pass ko band karne ke liye ham torch.no_grad() ka use kar sakte hai, isse ham gradients calculate nahi 
# karenge aur isse memory bhi bach jayegi.
# method 1
with torch.no_grad():
    z = w*x + b
```

```python
# method 2
w.requires_grad_(False)
b.requires_grad_(False)
```
