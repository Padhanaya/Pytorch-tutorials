# Inplace operations

```python
# when we add m + n a new tensor is created uses memory and space.
# agar hame n ko m me jodna h or ham ye bhi nhi chahte ki extra space use ho uske liye ham add_ function use karte h
m = torch.randint(size = (3,2), low = 1, high = 5, dtype = torch.float32)
n = torch.randint(size = (3,2), low = 1, high = 5, dtype = torch.float32)
print(m)
print(n)

# adding n in m
m.add_(n)
```

```python
m
```

```python
# similarly agar ham relu ke aage underscore (relu_) lage de to wo inplace operation karega
torch.relu_(m)
```

```python
# same for sigmoid
torch.sigmoid_(m)
```

# copy tensor

```python
a = torch.rand(3,2)
b = torch.rand(3,2)

b = a
```

```python
b
```

```python
a
```

```python
print(id(a))
print(id(b))

# both will show same memory id, means change in b cause change in a, which may be problem in future operations
```

```python
# Thus we uses clone method to make copy
a = torch.clone(b)
```

```python
print(id(a))
print(id(b))
# now both are using diff memory space
```
